```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pymongo
import requests
```

```python
plt.style.use('ggplot')
```

```python
# connecting to my mongodb and get data out
client = pymongo.MongoClient('mongodb://localhost:27017/')
# create database
db = client.movie_db
# create colletions
col_imdb = db['imbd_top250']
col_douban = db['douban_top250']
```

```python
# import data to dataframe
imdb_data = pd.DataFrame(list(col_imdb.find()))
```

# imdb Movie 

```python
# imdb top 250 rated
imdb_rated_s = imdb_data.groupby('rated').count()['movie'].sort_values(ascending=False)
```

```python
imdb_rated_s
```

```python
plt.pie(imdb_rated_s.values, autopct='%1.1f%%')
plt.legend(imdb_rated_s.index)
```

# douban Movie

```python
douban_data = pd.DataFrame(list(col_douban.find()))
douban_data.head()
```

```python
both_df = douban_data.merge(imdb_data, on='imdb_id', how='inner').drop(columns=['_id_x', '_id_y', 'year_x'])
```

```python
rank_df = both_df[['cn_name', 'movie', 'douban_rank', 'imdb_rank']]
```

```python
rank_df = rank_df.astype({'douban_rank': float, 'imdb_rank': float})
```

```python
rank_df.head()
```

```python
rank_df['avg_rank'] = (rank_df['douban_rank'] + rank_df['imdb_rank']) / 2
```

```python
sort_rank = rank_df.sort_values('avg_rank')
```

```python
sort_rank.plot(kind='line', x='movie', figsize=(10, 7))
```

```python
sort_rank.mean()
```

```python
sort_rank['dif_rank'] = sort_rank['imdb_rank'] - sort_rank['douban_rank']
```

```python
low_dif = sort_rank.sort_values('dif_rank', ascending=False)
#low_dif.loc[low_dif['dif_rank'] == 0]
```

```python
#low_dif.loc[low_dif['dif_rank'] > 0]
```

```python
#low_dif.loc[low_dif['dif_rank'] < 0].sort_values('dif_rank')
```

```python
both_rated_s = both_df.groupby('rated').count()['movie'].sort_values(ascending=False)
```

```python
plt.pie(both_rated_s.values, autopct='%1.1f%%')
plt.legend(both_rated_s.index)
```
